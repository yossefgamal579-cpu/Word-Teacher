```python
from textblob import Word
from googletrans import Translator

print("👋 Hello! I'm the Word Teacher.")
print("Type ONE word, and I will analyze it with NLP!")
print("------------------------------------------------")


while True:
  user_word =input("\n Enter a word or type 'quite' to stop: ").lower()
  if user_word == "quite":
    print(" Bye!")
    break

  # 1- Word started to analyze
  w = Word(user_word)

  # 2- check on spelling
  corrected = w.spellcheck()[0][0]

  # 3- word is lemmatized
  lematized =Word(corrected).lemmatize('v')

  translator = Translator()
  arabic_translation = translator.translate(user_word, dest='ar').text

  print("------------------------------------------------")
  print(f"\n🔎 NLP Analysis for '{user_word}':")
  print(f"✅ Corrected spelling: {corrected}")
  print(f"🌱 Lemma (base form): {lematized}")
  print(f"\n🌍 Arabic Translation for '{user_word}':")
  print(f"➡️ {arabic_translation}")
```
